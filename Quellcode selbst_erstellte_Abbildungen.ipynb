```python
metriken = ['RMSE', 'MAE', 'MSE', 'FCP']
bs_ib_values = [0.9959, 0.7751, 0.9919, 0.451]
bs_ub_values = [0.9944, 0.7679, 0.989, 0.6531]
wm_ib_values = [0.9274, 0.7102, 0.86, 0.639]
wm_ub_values = [0.9216, 0.7065, 0.8494, 0.6388]
colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3']

fig, ax = plt.subplots() #Quelle 4

ax.plot(metriken, bs_ib_values, marker='o', linestyle='-', color=colors[0], label='BS-IB') #Quelle 6
ax.plot(metriken, bs_ub_values, marker='o', linestyle='-', color=colors[1], label='BS-UB') #Quelle 6
ax.plot(metriken, wm_ib_values, marker='o', linestyle='-', color=colors[2], label='WM-IB') #Quelle 6
ax.plot(metriken, wm_ub_values, marker='o', linestyle='-', color=colors[3], label='WM-UB') #Quelle 6

ax.set_xlabel('Metriken') #Quelle 7
ax.set_ylabel('Werte') #Quelle 8

ax.set_ylim(0.4, 1.0) #Quelle 9
ax.legend() #Quelle 10

plt.show() # Quelle 5
```
